```python
evaluations = []

for i in range(2):  # só 2 exemplos pra começar
    question = input("Question: ")

    response_a = input("Response A: ")
    response_b = input("Response B: ")

    rel_a = input("Relevance A (1-5): ")
    rel_b = input("Relevance B (1-5): ")

    choice = input("Better response (A/B): ")
    justification = input("Why: ")

    data = {
        "question": question,
        "response_a": response_a,
        "response_b": response_b,
        "best": choice,
        "justification": justification
    }

    evaluations.append(data)

print(evaluations)
```

```python
import csv

with open("evaluation_dataset.csv", "w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=["question", "response_a", "response_b", "best", "justification"])

    writer.writeheader()

    for row in evaluations:
        writer.writerow(row)

print("Saved!")
```
